This is a file only for bond selection method of bucketing the bonds

```python
import pandas as pd
import numpy as np

# Load raw data

raw_df = pd.read_csv("bonds_master_all.csv")

print("Shape of raw data:", raw_df.shape)
print("\nColumns in raw data:")
print(raw_df.columns.tolist())

raw_df.head()
```

```python
# Extract relevant columns
df_required = raw_df[['date', 'Term', 'Coupon', 'Name', 'ISIN', 'Issue Date', 'Maturity Date', 'Close Price (% of par)']]
df_required = df_required.rename(columns={'date': 'Date'})

DATE = 'Date'
TERM = 'Term'
COUPON = 'Coupon'
NAME = 'Name'
ISIN = 'ISIN'
ISSUE_DATE = 'Issue Date'
MATURITY_DATE = 'Maturity Date'
CLOSE_PRICE = 'Close Price (% of par)'

df_required[DATE] = pd.to_datetime(df_required[DATE])
df_required[ISSUE_DATE] = pd.to_datetime(df_required[ISSUE_DATE])
df_required[MATURITY_DATE] = pd.to_datetime(df_required[MATURITY_DATE])
df_required.head()
```

```python
coupon_num = (
    df_required[COUPON].astype(str).str.replace('%', '', regex=False).astype(float)
)
# remove extra 0s
coupon_str = coupon_num.map(lambda x: f"{x:g}")

mon = df_required[MATURITY_DATE].dt.strftime('%b')   # Jan, Feb, Mar, ...
yy  = df_required[MATURITY_DATE].dt.strftime('%y')   # 26, 34, ...

# rename NAME
df_required[NAME] = "CAN " + coupon_str + " " + mon + " " + yy


print("Shape of required data:", df_required.shape)
df_required.head()
```

```python
# check if any required column has missing values
df_required.isna().sum()
```

```python
df_dates = df_required[[DATE, MATURITY_DATE]].copy()

df_dates['time_to_maturity'] = (
    (df_dates[MATURITY_DATE] - df_dates[DATE]).dt.days / 365
)

df_0_5 = df_dates[
    (df_dates['time_to_maturity'] >= 0) &
    (df_dates['time_to_maturity'] <= 5.5)
]

df_required_0_5 = df_required.loc[df_0_5.index].copy()

df_required_0_5['time_to_maturity'] = (
    (df_required_0_5[MATURITY_DATE] - df_required_0_5[DATE]).dt.days / 365
)

import numpy as np

bins = np.arange(0, 6, 0.5)   # 0, 0.5, 1, ..., 5
labels = [f"{bins[i]}–{bins[i+1]}Y" for i in range(len(bins)-1)]

df_required_0_5['ttm_bucket_0_5'] = pd.cut(
    df_required_0_5['time_to_maturity'],
    bins=bins,
    labels=labels,
    right=False,
    include_lowest=True
)
bond_count_half_year = (
    df_required_0_5
    .groupby('ttm_bucket_0_5')[ISIN]
    .nunique()
)

print(bond_count_half_year)

print(df_required_0_5[ISIN].nunique())

print(df_required_0_5)
```

```python
import numpy as np
import pandas as pd

# Assumes df_required already exists and dates are datetime
# Required constants: DATE, TERM, COUPON, NAME, ISIN, ISSUE_DATE, MATURITY_DATE, CLOSE_PRICE

# 1) Define the target window (auto from your dataset, but you can hardcode if you want)
window_dates = sorted(df_required[DATE].unique())
t0 = window_dates[0]              # earliest trading day (e.g., 2026-01-05)
t1 = window_dates[-1]             # latest trading day  (e.g., 2026-01-19)
expected_dates = set(window_dates)

print("Reference date t0:", pd.Timestamp(t0).date())
print("Window:", pd.Timestamp(t0).date(), "to", pd.Timestamp(t1).date())
print("Expected trading days in window:", len(expected_dates))

# 2) Work on a copy restricted to the window (safe)
df_win = df_required[df_required[DATE].isin(expected_dates)].copy()

# 3) Compute TTM at reference date t0 for each bond (unique by ISIN + Maturity Date)
#    Use maturity - t0 (constant across days)
bond_master = (
    df_win[[ISIN, NAME, TERM, COUPON, ISSUE_DATE, MATURITY_DATE]]
    .drop_duplicates()
    .copy()
)
bond_master['ttm_ref'] = (bond_master[MATURITY_DATE] - pd.Timestamp(t0)).dt.days / 365.0

# Keep only 0–5y at reference date
bond_master_0_5 = bond_master[(bond_master['ttm_ref'] >= 0) & (bond_master['ttm_ref'] <= 5.5)].copy()

# 4) Bucket into 0.5-year intervals
bins = np.arange(0, 5.5 + 0.5, 0.5)  # 0,0.5,...,5.0
labels = [f"{bins[i]:.1f}–{bins[i+1]:.1f}Y" for i in range(len(bins)-1)]

bond_master_0_5['ttm_bucket'] = pd.cut(
    bond_master_0_5['ttm_ref'],
    bins=bins,
    labels=labels,
    right=False,
    include_lowest=True
)

# 5) Completeness check per bond: do we have ALL expected trading days?
#    Key = (ISIN, Maturity Date) is safest (some ISIN issues can be tricky; usually ISIN unique)
obs_counts = (
    df_win.merge(bond_master_0_5[[ISIN, MATURITY_DATE]], on=[ISIN, MATURITY_DATE], how='inner')
         .groupby([ISIN, MATURITY_DATE])[DATE]
         .nunique()
         .rename('n_dates_observed')
         .reset_index()
)

bond_master_0_5 = bond_master_0_5.merge(obs_counts, on=[ISIN, MATURITY_DATE], how='left')
bond_master_0_5['n_dates_observed'] = bond_master_0_5['n_dates_observed'].fillna(0).astype(int)
bond_master_0_5['is_complete'] = bond_master_0_5['n_dates_observed'].eq(len(expected_dates))

# 6) Add one representative clean price for quick human inspection (price on t0 if exists)
#    (Optional but helpful for manual selection)
px_t0 = df_win[df_win[DATE] == pd.Timestamp(t0)][[ISIN, MATURITY_DATE, CLOSE_PRICE]].copy()
px_t0 = px_t0.rename(columns={CLOSE_PRICE: 'clean_price_t0'})

bond_master_0_5 = bond_master_0_5.merge(px_t0, on=[ISIN, MATURITY_DATE], how='left')

# 7) Final table for manual selection
manual_table = bond_master_0_5[[
    'ttm_bucket',
    'ttm_ref',
    ISIN,
    NAME,
    COUPON,
    TERM,
    ISSUE_DATE,
    MATURITY_DATE,
    'clean_price_t0',
    'n_dates_observed',
    'is_complete'
]].sort_values(['ttm_ref'])

# 8) Export:
#    - Sheet 1: All candidates (0–5y)
#    - Sheet 2: Only complete bonds (recommended pool)
#    - Extra sheets: one per bucket (optional, convenient for manual picking)
out_file = "Bond_Selection_0_5.5y_Bucketed.xlsx"

with pd.ExcelWriter(out_file) as writer:
    manual_table.to_excel(writer, sheet_name="All_0_5y", index=False)

    manual_table[manual_table['is_complete']].to_excel(writer, sheet_name="Complete_only", index=False)

    # one sheet per bucket
    for b, g in manual_table.groupby('ttm_bucket'):
        if pd.isna(b):
            continue
        sheet = str(b).replace('–', '-')  # Excel sheet name safe
        g.to_excel(writer, sheet_name=sheet[:31], index=False)

# print("Exported:", out_file)
# print("Complete bonds available:", manual_table['is_complete'].sum())

#then the bonds was anually selected
```
